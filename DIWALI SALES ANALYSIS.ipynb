```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="latin-1")
```

```python
df.head()
```

```python
df.shape
```

```python
df.size
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
#check for null vallues
df.isnull().sum()
```

```python
#drop null values as here according to the data null values are less
df.dropna(inplace=True)
```

```python
## change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df.columns
```

```python
df.rename(columns={"Marital_Status":"status"})
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

# EXPLORATORY DATA ANALYSIS

### GENDER

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sales_gen
```

```python
sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### Age

```python
ax=sns.countplot(x="Age Group",data=df,hue="Gender")
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
age_grp=pd.DataFrame(df[["Age Group","Gender","Amount"]].groupby("Age Group").value_counts())
age_grp
```

```python
sns.barplot(x = 'Age Group',y= 'Amount' ,data = age_grp)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

## State

```python
sns.countplot(x="State",data=df)
plt.xticks(rotation=90)
plt.show
```

```python
# total number of orders from top 10 states

ordr_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = ordr_state, x = 'State',y= 'Orders')
#orders placed by each state
```

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.barplot(x="State",y="Amount",data=sales_state)
plt.xticks(rotation=90)
plt.show
#amount spent by each state
```

```python
#From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

## marital status

```python
stats_grp=df.groupby(["Marital_Status"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
stats_grp
```

```python
sns.barplot(x="Marital_Status",y="Amount",data=stats_grp)
sns.set(rc={'figure.figsize':(6,5)})
#married purchased for more amount compared to singles 
```

```python
ordr_sts=df.groupby(["Marital_Status","Gender"],as_index=False)["Orders"].count().sort_values(by="Orders",ascending=False)
```

```python
sns.barplot(x="Marital_Status",y="Orders",hue="Gender",data=ordr_sts)
sns.set(rc={'figure.figsize':(6,5)})
#orders were also mostly done by married
```

```python
#married women made more shopping compared to 1
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

```python
#From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

# Product Category

```python
df["Product_Category"].unique()
```

```python
sns.countplot(x="Product_Category",data=df)
plt.xticks(rotation=90)
```

```python
pdt_cat=df.groupby(["Product_Category","Gender","Amount"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
pdt_cat
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics categor

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
